# Establish connections between keypoints

Define a principal function that draws the connections between the human skeleton. 
Link establishes the connection between all points. Here, 0-18 represents a feature point in a joint, 
for example, 0 represents the top of the head, 2,3 represents the left and right eyes, etc. 
For example, [0,2] represents the connection between feature point 0 and feature point 2.
In the loop, index1 and index2 represent each point of link. Use the POLT function to draw the connection between each
two points through a loop.

```python
def renderBones():
   link = [[0,2],[0,3],[2,5],[3,6],[1,5],[1,6]
             ,[5,11],[11,15],[6,12],[12,16]
             ,[3,8],[8,14],[14,17], [2,7],[7,13],[13,18]
             ,[1,4],[4,9],[4,10]]
   for l in link:
       index1,index2 = l[0],l[1]
       ax.plot([xs[index1],xs[index2]], [ys[index1],ys[index2]], [zs[index1],zs[index2]], linewidth=1, label=r'$z=y=x$')

```

```python
file="humanpose_data.txt"

nl=[]
with open (file) as f :
    for line in f:
        nl.extend([float(i) for i in line.split()])

xs=nl[0:19]
print(xs)
ys=nl[19:38]
print(ys) 
zs=nl[38:57]
print(zs) 
```
